```python
import nltk
from operator import itemgetter
import string
import re
from nltk.tokenize import sent_tokenize, word_tokenize



file = open(r"alice.txt")
text = file.read()
tokens = 0
wordcount = {} 

#regexs to remove puntuations 
reg = "[A-Za-z0-9]*"


wordcount['sent_start'] = 0
wordcount['sent_end'] = 0


#tokeniztion  and removal of unnecessary symbols
# also adding sent_start and sent_end at start and end of sentece respectively
sent = sent_tokenize(text)
for i in range(len(sent)): 
    words = word_tokenize(sent[i])
    sent[i] = 'sent_start '
    for word in words:
        word = word.lower()
        word = (re.match(reg,word)).group(0)
        if word =="":
            continue;
        tokens +=1
        if word not in wordcount:
            wordcount[word] = 1
        else:
            wordcount[word] += 1
        sent[i] = sent[i] + word +" "
    sent[i] = sent[i] + 'sent_end'
    wordcount['sent_start'] = wordcount['sent_start'] +1
    wordcount['sent_end'] = wordcount['sent_end'] +1
        
#dividing the dataset for traing(80%) and testing(20%) purpose
sent_train = sent[:int(0.8*len(sent))]
sent_test = sent[int(0.8*len(sent)):]
print("                                                   training dataset")
print(sent_train)
print("-------------------------------------------------------------------------------------")
print("                                                   testing dataset")
print(sent_test)
```

```python
#MLE for unigram
print("# for unigrams")
print("--------------",end = "\n\n")
prob_uni = {}
for key, value in reversed(sorted(wordcount.items(), key = itemgetter(1))):
    prob_uni[key] = value/tokens
    print("p(" , key ,') = ' , str(prob_uni[key]))
```

```python
#MLE for Bigram
from nltk.util import ngrams

print("#for bigram")
print("------------")

bigram_freq = {}
prob_Bi = {}
bigram_count = 0
print("#MLE for bigrams")
for sent in sent_train:
    bigrams=ngrams(nltk.word_tokenize(sent),2)
    for i in bigrams:
        bigram_count += 1
        if i not in bigram_freq:
            bigram_freq[i] = 1
        else:
            bigram_freq[i] += 1
            
for key, value in reversed(sorted(bigram_freq.items(), key = itemgetter(1))):
    prob_Bi[key] =  value/wordcount[key[0]]
    print("p(" , key[0],key[1] ,') = ' , str(prob_Bi[key]))
            

        
```

```python
#MLE for Trigram

trigram_freq = {}
trigram_count = 0
print("#MLE for trigrams")
print("------------------")
prob_TRI = {} 
for sent in sent_train:
    trigrams=ngrams(nltk.word_tokenize(sent),3)
    #print("error eror eror")
    for i in trigrams:
        trigram_count += 1
        if i not in trigram_freq:
            trigram_freq[i] = 1
        else:
            trigram_freq[i] += 1
#printing MLE
for key, value in reversed(sorted(trigram_freq.items(), key = itemgetter(1))):
    prob_TRI[key] = value/bigram_freq[(key[0],key[1])]
    print("p(" , key[0],key[1],key[2] ,') = ' , str(prob_TRI[key]))
```

```python
#MLE for Quadgram

quadgram_freq = {}
quadgram_count = 0
prob_QUAD = {}
print("#MLE for quadgrams")
print("------------------")
for sent in sent_train: 
    quad = nltk.word_tokenize(sent)
    if len(quad) < 4:
        continue
    quadgrams = ngrams(quad,4)
    #print()
    #print(nltk.word_tokenize(sent))
    for i in quadgrams:
        quadgram_count += 1
        if i not in quadgram_freq:
            quadgram_freq[i] = 1
        else:
            quadgram_freq[i] += 1
#print(quadgram_freq)
for key, value in reversed(sorted(quadgram_freq.items(), key = itemgetter(1))):
    prob_QUAD[key] = value/trigram_freq[(key[0],key[1],key[2])]   
    print("p(" , key[0],key[1],key[2],key[3] ,') = ' , str(prob_QUAD[key]))
```

```python
# function for checking prob of sentence using n gram model
import warnings
import math


#putting values of probabilty of all uni, bi,tri and quad in single dictionary
# as all have tupple keys and no of items in tupple is different, it can easily be distiquished 
prob = {}
prob.update(prob_uni)
prob.update(prob_Bi)
prob.update(prob_TRI)
prob.update(prob_QUAD)

#function for checking prob of sentence using n gram model
#the value of nmodel is used to find which model is to be used:
    # nmodel = 1 -> unigram
    #nmodel =2 -> bigram
    #nmodel = 3 -> trigram
    #nmodel = 4 -> quadgram
#isforsmootheddata is used later on when perplexity after smmothing is to calculated
def sent_prob(sentence,nmodel,isforsmootheddata,prob_Smmoothed):
    log_prob = 0
    nmodel = int(nmodel)
    token = nltk.word_tokenize(sentence)
    if nmodel > 1:
        token = ngrams(token,int(nmodel))
    
    warnings.filterwarnings('ignore')
    if isforsmootheddata == 1:
        for i in token:
            if i in prob_Smmoothed.keys():
                log_prob += math.log(prob_Smmoothed[i],2)
    else:    
        for i in token:
            log_prob += math.log(prob[i],2)
    return 2**log_prob

sentence = sent_train[150]
print("checking probabilty of sentence")
print("--------------------------------",end = "\n\n")
print(sentence,end ='\n\n')
#checking probabiltty of sentence using various model
p1 = sent_prob(sent_train[150],1,0,{})
print("probab of sentence in log  with base 2 using unigram model:=" , p1)
p2 = sent_prob(sent_train[150],2,0,{})
print("probab of sentence in log with base 2 using bygram modal :=" , p2)
p3 = sent_prob(sent_train[150],3,0,{})
print("probab of sentence in log with base 2 using trigram model :=" , p3)
p4 = sent_prob(sent_train[150],4,0,{})
print("probab of sentence in log with base 2 using quadgram model:=" , p4)
print("",end ="\n\n")

sentence = sent_train[100]
print(sentence,end ='\n\n')
#checking probabiltty of sentence using various model
p1 = sent_prob(sent_train[100],1,0,{})
print("probab of sentence in log  with base 2 using unigram model:=" , p1)
p2 = sent_prob(sent_train[100],2,0,{})
print("probab of sentence in log with base 2 using bygram modal :=" , p2)
p3 = sent_prob(sent_train[100],3,0,{})
print("probab of sentence in log with base 2 using trigram model :=" , p3)
p4 = sent_prob(sent_train[100],4,0,{})
print("probab of sentence in log with base 2 using quadgram model:=" , p4)
print("",end ="\n\n")

```

```python
print("Good Turing and finding value of d")
print("-------------------------------------")

fresDis = {}
good_count = {}
# to find how many word occur certain no of time
for key,value in bigram_freq.items():
    if value in fresDis:
        fresDis[value] += 2
    else:
        fresDis[value] = 1
#adding words which occur zero times
# by taking possible and bigrams which are present in courpus
fresDis[0] =   len(wordcount)**2 - len(bigram_freq)
print("Count c    Good Turing c*     difference")

for i in range(10):
    if (i+1)in fresDis and i in fresDis:
        c= (i+1)*fresDis[int(i+1)]/ fresDis[i]
        good_count[i] = c 
        print(i,"    ", c,"     " , abs(c-i))

print("The difference between c and c* lie nearly around 0.8")
print("=>value of d = 0.8")
```

```python
#finding perplexity using Good-turing
print("perplexity of test data using probab after Good Smoothing")
print("----------------------------------------------------")
print()
print('corresponding sentence has perplexity-->')
print("-----------------------------------------")
for sent in sent_test:
    #function call to find probab of sentence
    prob_log = 1
    token = nltk.word_tokenize(sent)
    token = ngrams(token,int(2))
    tokencount = 1    
    for key in token:
        # to cal no of bigram in given sentence
        tokencount += 1
        if key in bigram_freq:
            freq = bigram_freq[key] 
        else:
            freq = 0
        if freq in good_count.keys():
            #using log * becom +
            prob_log += math.log(good_count[freq]/bigram_count)
    if tokencount == 0 or 2**prob_log==0 :
        continue
    # taking antilog and reversing
    invprob = (1/2**prob_log)
    perplex = invprob**(1/tokencount)
    print( sent, " = ", perplex )
                   
```
